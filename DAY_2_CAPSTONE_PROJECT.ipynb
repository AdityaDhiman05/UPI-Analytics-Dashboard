```python

import pandas as pd
df = pd.read_csv('upi_data_enhanced.csv')
df.head()
```

```python
df['Month']=pd.to_datetime(df['Month'],format='%b-%y')
```

```python
df.sort_values('Month')
```

visualizing upi growth - volume growth

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(df['Month'], df['Volume (in Mn)'], marker='o', color='blue', label='Volume (Mn)')
plt.title('UPI Transaction Volume Growth Over Time')
plt.xlabel('Month')
plt.ylabel('Volume (in Million)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```

upi growth visualization - value growth

```python
plt.figure(figsize=(12,6))
plt.plot(df['Month'], df['Value (in Cr.)'], marker='o', color='green', label='Value (Cr.)')
plt.title('UPI Transaction Value Growth Over Time')
plt.xlabel('Month')
plt.ylabel('Value (Crores â‚¹)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(12,6))

color = 'tab:blue'
ax1.set_xlabel('Month')
ax1.set_ylabel('Volume (Mn)', color=color)
ax1.plot(df['Month'], df['Volume (in Mn)'], color=color, marker='o')
ax1.tick_params(axis='y', labelcolor=color)
ax1.tick_params(axis='x', rotation=45)

ax2 = ax1.twinx()  # second y-axis
color = 'tab:green'
ax2.set_ylabel('Value (Cr.)', color=color)
ax2.plot(df['Month'], df['Value (in Cr.)'], color=color, marker='s')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('UPI Volume and Value Growth Over Time')
plt.grid(True)
plt.show()
```
